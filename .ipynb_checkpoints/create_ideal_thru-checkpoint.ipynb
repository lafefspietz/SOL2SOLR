```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import os
```

```python
cold_ideals_2port = rf.read_all('3Kideals_2port')
frequency_object = cold_ideals_2port['ideal_port1_short'].frequency
fghz = cold_ideals_2port['ideal_port1_short'].f/1e9
```

```python
smatrix = []
for element in range(len(fghz)):
    localmatrix = np.identity(2) +  np.zeros((2,2))*1j
    smatrix.append(localmatrix)
ideal_thru = rf.Network(frequency=frequency_object, s=smatrix)
```

```python
ideal_thru.write_touchstone("3Kideals_2port/idealthru.s2p")
ideal_thru.write_touchstone("295Kideals_2port/idealthru.s2p")
```
