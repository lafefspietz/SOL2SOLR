```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import os
```

Notebook for taking a set of "ideals" from a 1 port calibration and turning it into 2 port ideals which assume zero transmission of reflection standards. This is not enough for full SOLR, but it's 3 of the 4 standards needed, and the R(reciprocal) can be measured in a single measurement of a single port-pair of any reciprocal device. I will now proceed to do this measurement at both room temperature and cold for 6 throughs to get some statistics of the over-determined system.  That is going to be the next measurement.

```python
# load all Touchstone files in "ideals" folder into scikit-rf
cold_ideals_1port = rf.read_all('3Kideals_1port')
warm_ideals_1port = rf.read_all('295Kideals_1port')
frequency_object = cold_ideals_1port['ideal_port1_short'].frequency
```

```python
cold_ideals_2port = {}
for ideal_1port in cold_ideals_1port:
    smatrix = []
    for element in cold_ideals_1port[ideal_1port].s:
        localmatrix = np.zeros((2,2)) + np.zeros((2,2))*1j
        localmatrix[0,0] = element[0,0]
        localmatrix[1,1] = element[0,0]
        smatrix.append(localmatrix)
    ideal = rf.Network(frequency=frequency_object, s=smatrix)
    cold_ideals_2port[ideal_1port] = ideal
```

```python
warm_ideals_2port = {}
for ideal_1port in warm_ideals_1port:
    smatrix = []
    for element in warm_ideals_1port[ideal_1port].s:
        localmatrix = np.zeros((2,2)) + np.zeros((2,2))*1j
        localmatrix[0,0] = element[0,0]
        localmatrix[1,1] = element[0,0]
        smatrix.append(localmatrix)
    ideal = rf.Network(frequency=frequency_object, s=smatrix)
    warm_ideals_2port[ideal_1port] = ideal
    
```

```python
for ideal_2port in cold_ideals_2port:
    cold_ideals_2port[ideal_2port].write_touchstone('3Kideals_2port/' + ideal_2port + ".s2p")
    
```

```python
for ideal_2port in warm_ideals_2port:
    warm_ideals_2port[ideal_2port].write_touchstone('295Kideals_2port/' + ideal_2port + ".s2p")
```
