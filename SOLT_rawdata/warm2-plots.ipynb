```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import json
```

```python
data = rf.read_all('warm2/')
fghz = data['short'].f/1e9
```

```python
plt.plot(fghz,data['short'].s_db[:,0,0],color="red")
plt.plot(fghz,data['open'].s_db[:,0,0],color="green")
plt.plot(fghz,data['load'].s_db[:,0,0],color="blue")
plt.grid()
plt.xlim(1,10)
plt.ylim(-40,5)
plt.ylabel('$S_{11} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['short','open','load'])
plt.title('Warm 2')
```

```python
plt.plot(fghz,data['short'].s_db[:,1,0],color="red")
plt.plot(fghz,data['open'].s_db[:,1,0],color="green")
plt.plot(fghz,data['load'].s_db[:,1,0],color="blue")
plt.grid()
plt.xlim(1,10)
plt.ylim(-120,5)
plt.ylabel('$S_{21} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['short','open','load'])
plt.title('Warm 2')
```

```python
plt.plot(fghz,data['port1'].s_db[:,1,0],color="red")
plt.plot(fghz,data['port2'].s_db[:,1,0],color="orange")
plt.plot(fghz,data['port3'].s_db[:,1,0],color="yellow")
plt.plot(fghz,data['port4'].s_db[:,1,0],color="green")
plt.plot(fghz,data['port5'].s_db[:,1,0],color="blue")
plt.plot(fghz,data['port6'].s_db[:,1,0],color="purple")

plt.grid()
plt.xlim(1,10)
plt.ylim(-25,1)
plt.ylabel('$S_{21} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['port 1','port 2','port 3','port 4','port 5','port 6'])
plt.title('Warm 2')
```
