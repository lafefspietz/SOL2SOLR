```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import json
```

```python
data = rf.read_all('warmup/')
fghz = data['5.141'].f/1e9
```

```python
for trace in data:
    print(trace)
```

```python
plt.plot(fghz,data['53.407'].s_db[:,1,0],color="red")
plt.plot(fghz,data['59.391'].s_db[:,1,0],color="orange")
plt.plot(fghz,data['64.483'].s_db[:,1,0],color="green")
plt.plot(fghz,data['69.025'].s_db[:,1,0],color="blue")
plt.plot(fghz,data['73.163'].s_db[:,1,0],color="purple")

plt.grid()
plt.xlim(1,10)
plt.ylim(-45,1)
plt.ylabel('$S_{21} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['T = 53 K','T = 59 K','T = 64 K','T = 69 K','T = 73 K'])
plt.title('Port 3 on Warm up')
```

```python
plt.plot(fghz,data['53.407'].s_db[:,1,0],color="red")
plt.grid()
plt.xlim(0.001,1)
plt.ylim(-100,1)
plt.ylabel('$S_{21} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.title('Port 3 T = 53.407')
```
