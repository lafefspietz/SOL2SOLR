```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import json
```

```python
data = rf.read_all('warm-port1-cable/')
fghz = data['short'].f/1e9
frequency_object = data['short'].frequency
ecal = rf.read_all('ecal-1port/')
```

```python
my_ideals_port1 = [\
    ecal['ideal_port1_short'],
    ecal['ideal_port1_open'],
    ecal['ideal_port1_load'],
]
```

```python
measured_short1 = rf.Network(frequency=frequency_object, s=data['short'].s[:,0,0])
measured_short2 = rf.Network(frequency=frequency_object, s=data['short'].s[:,1,1])
measured_open1 = rf.Network(frequency=frequency_object, s=data['open'].s[:,0,0])
measured_open2 = rf.Network(frequency=frequency_object, s=data['open'].s[:,1,1])
measured_load1 = rf.Network(frequency=frequency_object, s=data['load'].s[:,0,0])
measured_load2 = rf.Network(frequency=frequency_object, s=data['load'].s[:,1,1])

port1_1 = rf.Network(frequency=frequency_object, s=data['port1'].s[:,0,0])
port1_2 = rf.Network(frequency=frequency_object, s=data['port1'].s[:,1,1])
```

```python
my_measured_port1_1 = [\
    measured_short1,
    measured_open1,
    measured_load1,
]
my_measured_port1_2 = [\
    measured_short2,
    measured_open2,
    measured_load2,
]
```

```python
cal_port1_1 = rf.OnePort(\
        ideals = my_ideals_port1,
        measured = my_measured_port1_1,
        )
cal_port1_2 = rf.OnePort(\
        ideals = my_ideals_port1,
        measured = my_measured_port1_2,
        )
```

```python
cal_port1_1.run()
cal_port1_2.run()
```

```python
port1_corrected_1 = cal_port1_1.apply_cal(port1_1)
port1_corrected_2 = cal_port1_1.apply_cal(port1_2)
```

```python
plt.plot(fghz,port1_corrected_1.s_deg_unwrap[:,0,0],color='red')
plt.plot(fghz,port1_corrected_2.s_deg_unwrap[:,0,0],color='blue')
#plt.plot(fghz,-1020*fghz,color='purple')

plt.legend(['Cable','No Cable'])
plt.xlim(0.01,4)
plt.ylim(-6000,1000)
plt.grid()
plt.xlabel('Frequency [GHz]')
plt.ylabel('$S_{11}$ [deg]')
```

```python
plt.plot(fghz,port1_corrected_1.s_deg_unwrap[:,0,0],color='red',marker='+')
plt.plot(fghz,-1025*fghz,color='purple')

plt.legend(['corrected cable','line'])
plt.xlim(0.01,0.2)
plt.ylim(-200,10)
plt.grid()
plt.xlabel('Frequency [GHz]')
plt.ylabel('$S_{11}$ [deg]')
plt.title('line = -1025*fghz')
```

```python
smatrix = []
findex = 0
for element in range(len(fghz)):
    localmatrix = np.array([[0. + 0.j, np.exp(-1j*1025*fghz[findex]*np.pi/180)*1.+0.j],[ np.exp(-1j*1025*fghz[findex]*np.pi/180)*1.+0.j,0.+0.j]])
    smatrix.append(localmatrix)
    findex += 1
ideal_thru_phasedelay_double = rf.Network(frequency=frequency_object, s=smatrix)

smatrix = []
findex = 0
for element in range(len(fghz)):
    localmatrix = np.array([[0. + 0.j, np.exp(-1j*512.5*fghz[findex]*np.pi/180)*1.+0.j],[ np.exp(-1j*512.5*fghz[findex]*np.pi/180)*1.+0.j,0.+0.j]])
    smatrix.append(localmatrix)
    findex += 1
ideal_thru_phasedelay = rf.Network(frequency=frequency_object, s=smatrix)
```

```python
plt.plot(fghz,ideal_thru_phasedelay_double.s_deg_unwrap[:,1,0])
plt.plot(fghz,ideal_thru_phasedelay.s_deg_unwrap[:,1,0])
```
